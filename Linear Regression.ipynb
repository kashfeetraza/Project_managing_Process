```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python


dataframe = pd.read_csv('dataset.csv')
dataframe.head()
```

```python
x = np.array(dataframe[['registered']]).reshape(-1,1)
x
```

```python
y = dataframe["cnt"]
y
```

```python
# The corrected X is a single feature
x_plot = dataframe['temp']
y = dataframe['cnt']

plt.scatter(x_plot, y)
plt.xlabel('Normalized Temperature (temp)') # Label is for the single feature
plt.ylabel('Total Rental Count (cnt)')
plt.savefig('temp_vs_cnt_scatter.png')
```

```python
# Data splitting
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)

print(len(X_train))
print(len(X_test))
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
i = model.predict([[50]])
print(i)
```

```python
# Evaluate the model
acc = model.score(X_test,y_test)
print(acc)
```
